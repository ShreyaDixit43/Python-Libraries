```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years, runs)
```

```python
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Run Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years, kohli, label="Virat")
plt.plot(years, sehwag, label="Virendra Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

# Format strings

```python
plt.plot(years, kohli, 'ro--', label="Kohli")
plt.plot(years, sehwag, 'g^:', label="Sehwag")
```

# Color and line style arguments

```python
plt.plot(years, kohli, color='orange', label="Kohli")
plt.plot(years, sehwag, color= 'red', label="Sehwag")
plt.plot(years, runs, color='blue' , label="Tendulkar")
plt.legend()
```

```python
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot(years, kohli, linewidth=3, label="Kohli")
plt.plot(years, sehwag, linewidth=2, label="Sehwag")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
with plt.xkcd():
    plt.style.use('ggplot')
    plt.plot(years, kohli, linewidth=3, label="Kohli")
    plt.plot(years, sehwag, linewidth=2, label="Sehwag")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
